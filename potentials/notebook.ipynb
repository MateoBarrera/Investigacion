```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
 
print("File location using os.getcwd():", os.getcwd())
```

```python
file = pd.read_csv(
    "../recrusos/global/D1.1_POWER_Point_Monthly_Timeseries_2011_2021_LST.csv")
file.head()
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
T2M = file[file['PARAMETER'] == 'T2M']
T2M.head()
# gca stands for 'get current axis'
ax = plt.gca()
T2M.plot(kind='line', x='YEAR', y='ANN', ax=ax)
plt.title('MERRA-2 Temperature at 2 Meters (C)')

plt.subplot(1, 2, 2)
ALLSKY_KT = file[file['PARAMETER'] == 'ALLSKY_KT']
ALLSKY_KT.head()
# gca stands for 'get current axis'
ax = plt.gca()
ALLSKY_KT.plot(kind='line', x='YEAR', y='ANN', ax=ax)
plt.title('CERES SYN1deg All Sky Insolation Clearness Index (dimensionless)')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
ALLSKY_SFC_SW_DWN  = file[file['PARAMETER'] == 'ALLSKY_SFC_SW_DWN']
ALLSKY_SFC_SW_DWN .head()
# gca stands for 'get current axis'
ax = plt.gca()
ALLSKY_SFC_SW_DWN .plot(kind='line', x='YEAR', y='ANN', ax=ax)
plt.title('CERES SYN1deg All Sky Surface Shortwave Downward Irradiance (kW-hr/m^2/day)')

plt.subplot(1, 2, 2)
CLRSKY_SFC_SW_DWN = file[file['PARAMETER'] == 'CLRSKY_SFC_SW_DWN']
CLRSKY_SFC_SW_DWN.head()
# gca stands for 'get current axis'
ax = plt.gca()
CLRSKY_SFC_SW_DWN.plot(kind='line', x='YEAR', y='ANN', ax=ax)
plt.title(
    'CERES SYN1deg Clear Sky Surface Shortwave Downward Irradiance (kW-hr/m^2/day) ')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
PRECTOTCORR = file[file['PARAMETER'] == 'PRECTOTCORR']
PRECTOTCORR.head()
# gca stands for 'get current axis'
ax = plt.gca()
PRECTOTCORR.plot(kind='line', x='YEAR', y='ANN', ax=ax)
plt.title('MERRA-2 Precipitation Corrected (mm/day)')

plt.subplot(1, 2, 2)
WS10M = file[file['PARAMETER'] == 'WS10M']
WS10M.head()
# gca stands for 'get current axis'
ax = plt.gca()
WS10M.plot(kind='line', x='YEAR', y='ANN', ax=ax)
plt.title('MERRA-2 Wind Speed at 10 Meters (m/s)')
plt.show()
```

```python
file = pd.read_csv(
    "../recrusos/hydro/caudal_medio_mensual/Valle.csv.csv")
file.head()
```

```python
file_valle = file.filter(items=['CodigoEstacion','Municipio', 'IdParametro', 'Fecha', 'Valor'])
file_valle
```

```python
#Filtrar por municipio
file_jamundi = file_valle.loc[file_valle['Municipio']== 'Jamundí']

#Extraer las estaciones 
stations_jamundi = file_jamundi.drop_duplicates(subset=['CodigoEstacion'])['CodigoEstacion'].to_list()
stations_jamundi

for item in stations_jamundi:
  print(item)
  print(file_jamundi.loc[file_jamundi['CodigoEstacion']==item])
```
